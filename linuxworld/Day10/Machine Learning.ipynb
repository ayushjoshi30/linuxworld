```python
def set_age(row):
    Pclass=row[0]
    age=row[1]
    if numpy.isnan(age):
        if Pclass==1:
            return 38
        elif Pclass==2:
            return 29
        else:
            return 25
    else:
        return age
```

```python
dataset["Age"]=dataset[["Pclass","Age"]].apply(set_age,axis=1)
```
