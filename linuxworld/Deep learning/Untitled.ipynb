```python
from keras.models import Sequential 
```

```python
model=Sequential()
```

```python
from keras.layers import Dense
```

```python
model.get_config()
```

```python
model.add(Dense(units=4,input_shape=(13,),activation="relu",bias_initializer="zeros",kernel_initializer="zeros"))
```

```python
model.get_config()
```

```python
model.summary()
```

```python
model.add(Dense(units=4,activation="relu",bias_initializer="zeros",kernel_initializer="zeros"))
```

```python
y_cat
```

```python
model.add(Dense(units=3,activation="softmax",bias_initializer="zeros",kernel_initializer="zeros"))
```

```python
model.compile(loss="categorical_crossentropy")
```

```python
model.fit(x,y_cat,epochs=20)
```
